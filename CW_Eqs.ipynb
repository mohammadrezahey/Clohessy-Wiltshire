```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

omega = 1.0  # Mean motion
t = np.linspace(0, 10, 1000)

x0, y0, z0 = 1, 0.2, 0.5  # Initial positions
xdot0, ydot0, zdot0 = 0.1, 0.1, 0.0  # Initial velocities
Ux, Uy, Uz = 0.01, 0, 0.01  # Control accelerations

# Calculate the solutions
x = (x0 - (2 * ydot0) / omega) * np.cos(omega * t) + \
    (xdot0 / omega) * np.sin(omega * t) + \
    (2 * ydot0 / omega) * t + \
    Ux / omega**2 * (1 - np.cos(omega * t))

y = y0 * np.cos(omega * t) + \
    (ydot0 - 2 * xdot0) * np.sin(omega * t) + \
    Uy / omega**2 * (1 - np.cos(omega * t))

z = z0 * np.cos(omega * t) + \
    (zdot0 / omega) * np.sin(omega * t) + \
    Uz / omega**2 * (1 - np.cos(omega * t))

plt.figure(figsize=(14, 6))

plt.subplot(3, 1, 1)
plt.plot(t, x, label='x(t)')
plt.xlabel('Time [s]')
plt.ylabel('x [m]')
plt.title('Relative Motion in x Direction')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, y, label='y(t)')
plt.xlabel('Time [s]')
plt.ylabel('y [m]')
plt.title('Relative Motion in y Direction')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, z, label='z(t)')
plt.xlabel('Time [s]')
plt.ylabel('z [m]')
plt.title('Relative Motion in z Direction')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# 3D Plot of the Trajectory
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, label='Trajectory')
ax.set_xlabel('x [m]')
ax.set_ylabel('y [m]')
ax.set_zlabel('z [m]')
ax.set_title('3D Trajectory of Relative Motion')
ax.grid(True)
ax.legend()

plt.show()
```
